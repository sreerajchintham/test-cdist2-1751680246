```python

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

universities_df = pd.read_csv("/content/Universities.csv")

numerical_df = universities_df.drop(columns=["College Name", "State"])

full_records = numerical_df.dropna()
notfull_records = numerical_df[numerical_df.isnull().any(axis=1)]

scaler = StandardScaler()
complete_scaled = scaler.fit_transform(full_records)

k_means = KMeans(n_clusters=5, random_state=42, n_init=10)
full_records = full_records.copy()
full_records['Cluster'] = k_means.fit_predict(complete_scaled)

tufts_record = universities_df[universities_df['College Name'].str.contains('Tufts University', case=False)]
tufts_record = tufts_record.drop(columns=["College Name", "State"])
tufts_data = tufts_record.iloc[0]

nonnull_columns = tufts_data[tufts_data.notnull()].index

small_scaler = StandardScaler()
subset_scaled = small_scaler.fit_transform(complete_records[nonnull_columns])
tufts_normalized = small_scaler.transform(pd.DataFrame([tufts_data[nonnull_columns]], columns=nonnull_columns))

distances = cdist(tufts_normalized, subset_scaled)
closest_index = distances.argmin()

closest_cluster_label = full_records.iloc[closest_index]['Cluster']

cluster_means = full_records[full_records['Cluster'] == closest_cluster_label].mean()

tufts_completed = tufts_data.copy()
for col in tufts_data.index:
    if pd.isnull(tufts_data[col]):
        tufts_completed[col] = cluster_means[col]

print(tufts_completed)

```
